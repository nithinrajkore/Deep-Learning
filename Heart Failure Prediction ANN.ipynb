```python
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# load dataset
dataset = pd.read_csv('heart_failure_clinical_records_dataset.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Builing an ANN

```python
# Builing an ANN
ann = tf.keras.models.Sequential()
```

```python
# Adding input layer and one hidden layer
ann.add(tf.keras.layers.Dense(units = 12, activation = 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units = 12, activation = 'relu'))
```

```python
# add on output layer
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

### Training an ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = [tf.keras.metrics.BinaryAccuracy(), tf.keras.metrics.FalseNegatives()])
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 350)
```

### making predictions and evaluating model

```python
ann.predict(sc.transform([[4.20e+01, 1.00e+00, 8.60e+01, 0.00e+00, 3.50e+01, 0.00e+00,
       3.65e+05, 1.10e+00, 1.39e+02, 1.00e+00, 1.00e+00, 2.01e+02]])) 
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
# print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

### Evaluating on train set

```python
y_train_pred = ann.predict(X_train)
y_train_pred = (y_train_pred > 0.5)
```
